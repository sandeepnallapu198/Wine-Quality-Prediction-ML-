```python
data['quality'] = data['quality'].map({3:0,4:1,5:2,6:3,7:4,8:5})
data
```

```python
x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 0)
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler

SC = StandardScaler()
x_train = SC.fit_transform(x_train)
x_test = SC.transform(x_test)
```

```python
from lazypredict.Supervised import LazyClassifier
clf = LazyClassifier()
models,predictions = clf.fit(x_train,x_test,y_train,y_test)
models
```

```python
plt.figure(figsize = (10,5))
sns.barplot(x = models.index, y = 'Accuracy', data = models )
plt.xticks(rotation = 90)
plt.show
```

```python
plt.figure(figsize = (5,10))
sns.barplot(x = 'Accuracy', y = models.index, data = models)
plt.show()
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(x_train,y_train)
pred = model.predict(x_test)
print(accuracy_score(y_test,pred)*100)
```

```python
print(classification_report(y_test,pred))
```

```python
sns.heatmap(confusion_matrix(y_test,pred),annot  = True)
```

```python
from sklearn.model_selection import cross_val_score

score = cross_val_score(model,x_train,y_train,cv = 10,scoring = 'accuracy')
print('mean_score',(score.mean()))
print('std_score',(score.std()))
```
